A partir do dataframe oficial da OMS é possível se ter visão a dados subdivididos nas seguintes colunas:

- *Data_Reported* | Data Atualização
- *Country_Code* | Código do país segundo controle da OMS
- *Country* | País
- *WHO_region* | Região segundo controle da OMS
- *New_cases* | Nº de Novos casos
- *Cumulative_cases* | Nº de  novos casos acumulados
- *New_deaths* | Nº de mortos confirmados
- *Cumulative_deaths* | Nº de mortos acumulados

## <font color="#1c4587">Análise de Dados & Questões</font>

```python
recorte_colunas = df_baseWHO.loc[0:colunas_totais[0], 'Data Atualização': 'Total Mortes'] # separação colunas em pt-br
data_recente = recorte_colunas.loc[len(df_baseWHO)-1, 'Data Atualização'] # recorte data mais recente do relatório da OMS
```

### **Análises Gerais**

#### **Evolução Covid**

#### **1. Total Número de casos X Continente:**

```python
casos_america = casos_europa= casos_africa = casos_asia = casos_oceania = casos_outros = 0

for i in range(0, len(recorte_colunas)):
  if recorte_colunas.loc[i, 'Data Atualização'] == data_recente and recorte_colunas.loc[i, 'Continente'] == 'América':
    casos_america += recorte_colunas.loc[i, 'Total Casos']
  elif recorte_colunas.loc[i, 'Data Atualização'] == data_recente and recorte_colunas.loc[i, 'Continente'] == 'Europa':
    casos_europa += recorte_colunas.loc[i, 'Total Casos']
  elif recorte_colunas.loc[i, 'Data Atualização'] == data_recente and recorte_colunas.loc[i, 'Continente'] == 'África':
    casos_africa += recorte_colunas.loc[i, 'Total Casos']
  elif recorte_colunas.loc[i, 'Data Atualização'] == data_recente and recorte_colunas.loc[i, 'Continente'] == 'Ásia':
    casos_asia += recorte_colunas.loc[i, 'Total Casos']
  elif recorte_colunas.loc[i, 'Data Atualização'] == data_recente and recorte_colunas.loc[i, 'Continente'] == 'Oceania':
    casos_oceania += recorte_colunas.loc[i, 'Total Casos']
  elif recorte_colunas.loc[i, 'Data Atualização'] == data_recente and recorte_colunas.loc[i, 'Continente'] == 'Outros':
    casos_outros += recorte_colunas.loc[i, 'Total Casos']
```

```python
continentes = ['Outros', 'África', 'Oceania', 'América', 'Europa','Ásia']
casos_totais = [casos_outros, casos_africa, casos_oceania, casos_america, casos_europa, casos_asia]

legenda = [f"{casos_outros:,}", f"{casos_africa:,}", f"{casos_oceania:,}", f"{casos_america:,}", f"{casos_europa:,}", f"{casos_asia:,}"]
cores = ['#D5F22C', '#E4AD44', '#EB8B50', '#F2685C', '#7A4C61', '#023066']

casos_continentes = plt.barh(continentes, casos_totais, label=legenda, color=cores)
plt.title('Total Nº de Casos X Continentes')
plt.xlim(0, max(casos_totais)+ 50000000)
plt.legend()
plt.show()
```

```python
filtragem_datas = recorte_colunas[recorte_colunas['Data Atualização']== data_recente] # Filtragem por data mais recente
filtragem_casos_totais = filtragem_datas[filtragem_datas['Total Casos']<= max(recorte_colunas['Total Casos'])] # filtragem por número de casos totais
classificacao = filtragem_casos_totais.sort_values(by='Total Casos', ascending=False) # Top 10 países com maior número de casos, ordenando dataframe com os maiores valores.
selecao_linhas = classificacao[0:10]
display(selecao_linhas[['Data Atualização', 'Continente', 'País', 'Cód. País', 'Total Casos']])
```
